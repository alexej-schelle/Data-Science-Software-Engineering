```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cross_decomposition import PLSRegression
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Beispieldaten erzeugen (multivariate Regression)
input_data, output_data = make_regression(n_samples=200, n_features=100, n_targets=1,
                       noise=0.5, random_state=42)

# Train/Test-Split
input_data_train, input_data_test, output_data_train, output_data_test = train_test_split(input_data, output_data, test_size=0.2, random_state=42)

# PLSRegression-Modell (2 Komponenten)
pls = PLSRegression(n_components=2)
pls.fit(input_data_train, output_data_train)

# Vorhersagen
output_data_pred = pls.predict(input_data_test)

# Bewertung
r2 = r2_score(output_data_test, output_data_pred)
print(f"RÂ²-Score: {r2:.2f}")

# Scatterplot der Vorhersagen
plt.figure(figsize=(8, 5))
plt.scatter(output_data_test, output_data_pred, color='navy', edgecolors='k', alpha=0.7)
plt.plot([output_data_test.min(), output_data_test.max()], [output_data_test.min(), output_data_test.max()], 'r--', lw=2)
plt.xlabel("Echte Werte")
plt.ylabel("Vorhergesagte Werte")
plt.title("PLSRegression (Cross Decomposition)")
plt.grid(True)
plt.show()
```
