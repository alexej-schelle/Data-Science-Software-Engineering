```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Erstelle 2D-Daten mit 'moon'-Muster
input_data, output_data = make_moons(n_samples=300, noise=0.1, random_state=42)

# Scatter-Plot der rohen Moon-Daten vor dem Clustering
plt.figure(figsize=(8, 6))
plt.scatter(input_data[:, 0], input_data[:, 1], c='gray', edgecolors='k', s=50, alpha=0.7)
plt.title("Rohdaten der Moon-Daten (Vor dem Clustering)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()

# Hierarchisches Clustering mit AgglomerativeClustering
clustering = AgglomerativeClustering(n_clusters=4, linkage='ward')
output_data_pred = clustering.fit_predict(input_data)

# Scatter-Plot der Cluster nach dem Hierarchischen Clustering
plt.figure(figsize=(8, 6))
plt.scatter(input_data[:, 0], input_data[:, 1], c=output_data_pred, cmap='viridis', edgecolors='k', s=50)
plt.title("Hierarchisches Clustering der Moon-Daten")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()

# Dendrogramm erstellen (optional) zur Visualisierung der Cluster-Hierarchie
Z = linkage(input_data, 'ward')
plt.figure(figsize=(10, 6))
dendrogram(Z)
plt.title("Dendrogramm fr Hierarchisches Clustering")
plt.xlabel("Index der Datenpunkte")
plt.ylabel("Abstand")
plt.show()
```
