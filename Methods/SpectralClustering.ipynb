```python
import os

# Setze die Anzahl der Threads für OMP (OpenMP)
os.environ["OMP_NUM_THREADS"] = "2"  # Oder eine andere Zahl, je nachdem, wie viele Threads du nutzen möchtest

# Setze die maximale Anzahl der CPU-Kerne, die `joblib` verwenden soll
os.environ["LOKY_MAX_CPU_COUNT"] = "2"  # Beispiel: auf 2 Kerne setzen

import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
from sklearn.cluster import SpectralClustering

# Erstelle 2D-Daten mit 'moon'-Muster
input_data, output_data = make_moons(n_samples=3000, noise=0.1, random_state=42)

# Scatter-Plot der rohen Moon-Daten vor dem Clustering
plt.figure(figsize=(8, 6))
plt.scatter(input_data[:, 0], input_data[:, 1], c='gray', edgecolors='k', s=50, alpha=0.7)
plt.title("Rohdaten der Moon-Daten (Vor dem Clustering)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()

# Spektrales Clustering
clustering = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', random_state=42)
output_data_pred = clustering.fit_predict(input_data)

# Scatter-Plot der Cluster nach dem Clustering
plt.figure(figsize=(8, 6))
plt.scatter(input_data[:, 0], input_data[:, 1], c=output_data_pred, cmap='viridis', edgecolors='k', s=50)
plt.title("Spektrales Clustering der Moon-Daten")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```
