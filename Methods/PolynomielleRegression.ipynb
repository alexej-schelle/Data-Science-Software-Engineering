```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# Beispiel-Daten generieren (nichtlinear)
np.random.seed(0)

input_data = np.sort(np.random.rand(300) * 6).reshape(-1, 1)
output_data = np.cos(input_data).ravel() + 0.3 * np.random.randn(300)

# Polynomgrad definieren
degree = 4  # Du kannst z.B. 2, 3, 5 ausprobieren

# Pipeline: Polynomielle Transformation + Lineare Regression
model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
model.fit(input_data, output_data)

# Vorhersagebereich erstellen
input_data_plot = np.linspace(0, 6, 300).reshape(-1, 1)
output_data_plot = model.predict(input_data_plot)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(input_data, output_data, color='blue', label='Trainingsdaten')
plt.plot(input_data_plot, output_data_plot, color='red', linewidth=2, label=f'Polynomielle Regression (Grad {degree})')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Polynomielle Regression")
plt.legend()
plt.grid(True)
plt.show()
```
