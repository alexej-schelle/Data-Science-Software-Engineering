```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.kernel_ridge import KernelRidge

# Beispiel-Daten (nichtlinear)
np.random.seed(0)

input_data = np.sort(5 * np.random.rand(100, 1), axis=0)
output_data = np.sin(input_data).ravel() + np.random.normal(0, 0.2, input_data.shape[0])

# Kernel Ridge Regression mit RBF (Radial Basis Function)
model = KernelRidge(kernel='rbf', alpha=1.0, gamma=0.5)
model.fit(input_data, output_data)

# Interpolationsbereich
input_data_plot = np.linspace(0, 5, 500).reshape(-1, 1)
output_data_plot = model.predict(input_data_plot)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(input_data, output_data, color='blue', s=30, label='Trainingsdaten')
plt.plot(input_data_plot, output_data_plot, color='red', linewidth=2, label='Kernel Ridge (RBF)')
plt.xlabel("Eingabedaten")
plt.ylabel("Ausgabedaten")
plt.title("Kernel Ridge Regression mit RBF-Kernel")
plt.legend()
plt.grid(True)
plt.show()
```
