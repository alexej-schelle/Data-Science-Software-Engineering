```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

# Generate sample data
np.random.seed(0)
input_data = 2 * np.random.rand(100, 1)  # 100 random numbers between 0 and 2
output_data = 4 + 3 * input_data + np.random.randn(100, 1)  # Linear relation with some noise

# Create and train the linear regression model
model = LinearRegression()
model.fit(input_data, output_data)

# Print the learned parameters
print(f"Intercept: {model.intercept_[0]:.2f}")
print(f"Slope: {model.coef_[0][0]:.2f}")

# Make predictions
input_data_new = np.array([[0], [2]])
output_data_predict = model.predict(input_data_new)

# Plot the data and the regression line
plt.scatter(input_data, output_data, color='blue', label='Datengrundlage')
plt.plot(input_data_new, output_data_predict, color='red', linewidth=2, label='Regeressionsgerade')
plt.xlabel("Datenimport")
plt.ylabel("Datenexport")
plt.title("Beispiel f√ºr Lineare Regression")
plt.legend()
plt.grid(True)
plt.show()
```
