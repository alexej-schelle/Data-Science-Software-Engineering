```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression, Ridge

# Daten generieren
np.random.seed(0)
input_data = 2 * np.random.rand(100, 1)
output_data = 4 + 3 * input_data.ravel() + np.random.randn(100)

# Ridge Regression trainieren
ridge_model = Ridge(alpha=1.0)  # alpha ist der Regularisierungsparameter
ridge_model.fit(input_data, output_data)

# Vergleich: normale lineare Regression
linear_model = LinearRegression()
linear_model.fit(input_data, output_data)

# Vorhersagebereich
input_data_range = np.linspace(0, 2, 100).reshape(-1, 1)
output_data_ridge_pred = ridge_model.predict(input_data_range)
output_data_linear_pred = linear_model.predict(input_data_range)

# Modellparameter ausgeben
print(f"Ridge Intercept: {ridge_model.intercept_:.2f}")
print(f"Ridge Slope: {ridge_model.coef_[0]:.2f}")
print(f"Linear Intercept: {linear_model.intercept_:.2f}")
print(f"Linear Slope: {linear_model.coef_[0]:.2f}")

# Plotten
plt.scatter(input_data, output_data, color='blue', label='Daten')
plt.plot(input_data_range, output_data_linear_pred, color='green', label='Lineare Regression')
plt.plot(input_data_range, output_data_ridge_pred, color='red', linestyle='--', label='Ridge Regression')
plt.xlabel("Inputdaten")
plt.ylabel("Outputdaten")
plt.title("Vergleich: Lineare vs. Ridge Regression")
plt.legend()
plt.grid(True)
plt.show()
```
