```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay

# Iris-Datensatz laden
iris = load_iris()
input_data, output_data = iris.data, iris.target

# Train/Test-Split
input_data_train, input_data_test, output_data_train, output_data_test = train_test_split(input_data, output_data, test_size=0.3, random_state=42)

# Decision Tree Classifier trainieren
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(input_data_train, output_data_train)

# Vorhersagen und Genauigkeit
output_data_pred = clf.predict(input_data_test)
print(f"Genauigkeit: {accuracy_score(output_data_test, output_data_pred):.2f}")

# Baum visualisieren
plt.figure(figsize=(12, 6))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Entscheidungsbaum für Iris-Datensatz")
plt.show()

# Konfusionsmatrix
ConfusionMatrixDisplay.from_predictions(output_data_test, output_data_pred, display_labels=iris.target_names)

plt.title("Konfusionsmatrix")
plt.grid(False)
plt.show()
```
