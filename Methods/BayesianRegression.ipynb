```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import BayesianRidge

# Beispiel-Daten generieren
np.random.seed(0)
input_data = np.linspace(0, 6, 30).reshape(-1, 1)
output_data = np.sin(input_data).ravel() + np.random.normal(0, 0.3, input_data.shape[0])  # Sinus mit Rauschen

# Bayesian Ridge Regression trainieren
model = BayesianRidge()
model.fit(input_data, output_data)

# Vorhersagen + Unsicherheiten fr dichte X-Werte (Interpolation)
input_data_test = np.linspace(0, 6, 300).reshape(-1, 1)
output_data_mean, output_data_std = model.predict(input_data_test, return_std=True)

# Plot: Daten, Vorhersage, Unsicherheitsband
plt.figure(figsize=(10, 6))
plt.scatter(input_data, output_data, color='blue', label='Trainingsdaten')
plt.plot(input_data_test, output_data_mean, color='red', label='Bayesian Regression')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Bayesian Ridge Regression mit Unsicherheitsband")
plt.legend()
plt.grid(True)
plt.show()
```
