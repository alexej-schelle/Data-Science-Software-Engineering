```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
from sklearn.datasets import make_classification

# 3D-Daten generieren
input_data, output_data = make_classification(n_samples=30000, n_features=5, n_informative=3,
                           n_redundant=0, n_clusters_per_class=1, n_classes=2, random_state=42)

# PCA-Objekt erstellen  Ziel: 2 Dimensionen
pca = PCA(n_components=2)
input_data_pca = pca.fit_transform(input_data)

# Plot: Originaldaten (projiziert in 2D)
plt.figure(figsize=(8, 6))
plt.scatter(input_data_pca[:, 0], input_data_pca[:, 1], c=output_data, cmap='coolwarm', edgecolors='k', alpha=0.7)
plt.title("PCA: Dimensionsreduktion von 3D  2D")
plt.xlabel("Hauptkomponente 1")
plt.ylabel("Hauptkomponente 2")
plt.grid(True)
plt.show()

# Erklrte Varianz anzeigen
print("Erklrte Varianz pro Komponente:", pca.explained_variance_ratio_)
print("Gesamte erklrte Varianz:", np.sum(pca.explained_variance_ratio_))
```
