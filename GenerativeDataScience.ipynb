```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

# Load dataset
file_path = 'Energy_Consumption_Dataset.csv'  # Replace with your path
df = pd.read_csv(file_path)

# Set seaborn style
sns.set(style="whitegrid")

# ----- Categorical Feature Distributions -----
categorical_cols = ['Month', 'Hour', 'DayOfWeek', 'Holiday', 'HVACUsage', 'LightingUsage']

plt.figure(figsize=(20, 20))
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(3, 2, i)
    sns.countplot(data=df, x=col, palette="Set2")
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ----- Numerical Feature Distributions -----
numerical_cols = ['Temperature', 'Humidity', 'SquareFootage', 'Occupancy', 'RenewableEnergy', 'EnergyConsumption']

plt.figure(figsize=(20, 20))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 2, i)
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Number of samples
n_samples = 1000

# Categorical features (uniform distribution of choices)
months = np.random.choice(range(1, 13), size=n_samples)                # 1 to 12
hours = np.random.choice(range(0, 24), size=n_samples)                # 0 to 23
days_of_week = np.random.choice(range(1, 8), size=n_samples)          # 1 to 7
holidays = np.random.choice(['Yes', 'No'], size=n_samples)           # Yes/No
hvac_usage = np.random.choice(['On', 'Off'], size=n_samples)         # On/Off
lighting_usage = np.random.choice(['On', 'Off'], size=n_samples)     # On/Off

# Numerical features (uniform distribution)
temperature = np.random.uniform(15, 35, size=n_samples)               # e.g., degrees Celsius
humidity = np.random.uniform(30, 80, size=n_samples)                  # %
square_footage = np.random.uniform(500, 5000, size=n_samples)         # ft
occupancy = np.random.uniform(0, 100, size=n_samples)                 # number of people
renewable_energy = np.random.uniform(0, 1, size=n_samples)           # kWh from renewables

# Energy Consumption (normally distributed)
energy_consumption = np.random.normal(loc=75, scale=20, size=n_samples)  # mean=300, std=50
energy_consumption = np.clip(energy_consumption, 0, None)  # no negative energy use

# Combine into DataFrame
df = pd.DataFrame({
    'Month': months,
    'Hour': hours,
    'DayOfWeek': days_of_week,
    'Holiday': holidays,
    'HVACUsage': hvac_usage,
    'LightingUsage': lighting_usage,
    'Temperature': temperature,
    'Humidity': humidity,
    'SquareFootage': square_footage,
    'Occupancy': occupancy,
    'RenewableEnergy': renewable_energy,
    'EnergyConsumption': energy_consumption
})

# Save to CSV
df.to_csv("Energy_Consumption_Dataset_AI.csv", index=False)

print("Synthetic dataset saved as 'Energy_Consumption_Dataset_AI.csv'")
```
